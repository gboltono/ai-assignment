Prompt: Create a percentage plot to show the percentage of fraudulent transactions when a PIN was used or not.

```python
# Compute percentage distribution of fraud for each PIN usage category
fraud_dist = (
    data_clean
    .groupby(['used_pin_number', 'fraud'])
    .size()
    .groupby(level=0, group_keys=False)
    .apply(lambda x: x / x.sum() * 100)
    .reset_index(name='percent')
)

# Pivot for stacked plotting
fraud_pivot = fraud_dist.pivot(index='used_pin_number', columns='fraud', values='percent')

# Plot 100% stacked bar chart
fraud_pivot.plot(
    kind='bar',
    stacked=True,
    color=['skyblue', 'salmon'],
    figsize=(7,5)
)

plt.title('Fraud Distribution by PIN Usage (100% Stacked)', fontsize=14)
plt.xlabel('Used PIN Number (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Fraud', labels=['Legit (0)', 'Fraudulent (1)'])
plt.ylim(0, 100)
plt.show()
```

Prompt: Now, put the count plot and the percentage plot in the same cell, so that they show up on top of each other

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define a consistent color palette
colors = ['#4C72B0', '#DD8452']  # blue for legit (0), orange for fraud (1)

# Create a figure with two rows
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# --- 1 Countplot: Fraudulent vs. Non-Fraudulent Transactions by PIN Usage ---
sns.countplot(
    data=data_clean,
    x='used_pin_number',
    hue='fraud',
    palette=colors,
    ax=axes[0]
)
axes[0].set_title('Fraudulent vs. Non-Fraudulent Transactions by PIN Usage', fontsize=14)
axes[0].set_xlabel('Used PIN Number (0 = No, 1 = Yes)', fontsize=12)
axes[0].set_ylabel('Transaction Count', fontsize=12)
axes[0].legend(title='Fraud', labels=['Legit (0)', 'Fraudulent (1)'])

# --- 2 Percentage (100% Stacked) Plot ---
fraud_dist = (
    data_clean
    .groupby(['used_pin_number', 'fraud'])
    .size()
    .groupby(level=0, group_keys=False)
    .apply(lambda x: x / x.sum() * 100)
    .reset_index(name='percent')
)

fraud_pivot = fraud_dist.pivot(index='used_pin_number', columns='fraud', values='percent')

fraud_pivot.plot(
    kind='bar',
    stacked=True,
    color=colors,
    ax=axes[1],
    legend=False
)

axes[1].set_title('Fraud Distribution by PIN Usage (100% Stacked)', fontsize=14)
axes[1].set_xlabel('Used PIN Number (0 = No, 1 = Yes)', fontsize=12)
axes[1].set_ylabel('Percentage (%)', fontsize=12)
axes[1].set_ylim(0, 100)
axes[1].legend(title='Fraud', labels=['Legit (0)', 'Fraudulent (1)'])

plt.tight_layout()
plt.show()
```

Promp: Now, create similar graphs for the online_order column: a count plot and a percentage plot to show the amount of fraudulent orders, for orders that were placed online. Put them in the same cell, on top of each other

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Reuse the same color palette
colors = ['#008000', '#FF0000']  # green for legit (0), red for fraud (1)

# Create figure with two rows
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# --- 1 Countplot: Fraudulent vs. Non-Fraudulent Transactions by Online Order ---
sns.countplot(
    data=data_clean,
    x='online_order',
    hue='fraud',
    palette=colors,
    ax=axes[0]
)
axes[0].set_title('Fraudulent vs. Non-Fraudulent Transactions by Online Order', fontsize=14)
axes[0].set_xlabel('Online Order (0 = No, 1 = Yes)', fontsize=12)
axes[0].set_ylabel('Transaction Count', fontsize=12)
axes[0].legend(title='Fraud', labels=['Legit (0)', 'Fraudulent (1)'])

# --- 2 Percentage (100% Stacked) Plot ---
fraud_dist_online = (
    data_clean
    .groupby(['online_order', 'fraud'])
    .size()
    .groupby(level=0, group_keys=False)
    .apply(lambda x: x / x.sum() * 100)
    .reset_index(name='percent')
)

fraud_pivot_online = fraud_dist_online.pivot(index='online_order', columns='fraud', values='percent')

fraud_pivot_online.plot(
    kind='bar',
    stacked=True,
    color=colors,
    ax=axes[1],
    legend=False
)

axes[1].set_title('Fraud Distribution by Online Order (100% Stacked)', fontsize=14)
axes[1].set_xlabel('Online Order (0 = No, 1 = Yes)', fontsize=12)
axes[1].set_ylabel('Percentage (%)', fontsize=12)
axes[1].set_ylim(0, 100)
axes[1].legend(title='Fraud', labels=['Legit (0)', 'Fraudulent (1)'])

plt.tight_layout()
plt.show()
```
